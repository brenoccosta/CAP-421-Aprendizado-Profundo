```python
import rasterio
import matplotlib.pyplot as plt

forest_mask_path = "./Data/LULC/Forest Cover Mask 2023.tif"
latent_mosaic_path = "./Data/Images/Latent Mosaic/Latent_Mosaic.tif"

with rasterio.open(forest_mask_path) as src:
  mask = src.read(1)

with rasterio.open(latent_mosaic_path) as src:
  mosaic = src.read(1)

forest_values = mosaic[mask == 1]
other_values = mosaic[mask == 0]

plt.suptitle("Histograma do espaço latente")
plt.title("Distribuição entre florestas e não florestas")
plt.hist(forest_values, bins=50, label="Florestas")
plt.hist(other_values, bins=50, label="Outros")
plt.xlabel("Valores do espaço latente")
plt.ylabel("Frequência")
plt.legend()
plt.show()
```
