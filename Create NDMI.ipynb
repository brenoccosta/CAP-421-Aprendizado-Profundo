```python
def get_ndmi(b08, b11):
  return 1e4 * np.divide((b08 - b11), (b08 + b11), where=b08 != 0)

import os
import rasterio
import numpy as np

imgs_dir = "./Data/Images"
walkthrough = os.walk(imgs_dir)
output_file = "NDMI.tif"

for root, folders, files in walkthrough:
  if files:
    print(f"[DEBUG] {root} ...", end=" ")
    for file in files:
      
      isb08 = "B08" in file
      if isb08:
        with rasterio.open(os.path.join(root, file)) as src:
          b08 = src.read(1)
          out_meta = src.meta
      
      isb11 = "B11" in file
      if isb11:
        with rasterio.open(os.path.join(root, file)) as src:
          b11 = src.read(1)
      
    ndmi = get_ndmi(b08, b11)
    
    with rasterio.open(os.path.join(root, output_file), "w", **out_meta) as dst:
      dst.write(ndmi, 1)
    print(f"{output_file}")
```
