```python
import os
import rasterio
from glob import glob
from rasterio.merge import merge

patches_dir = "./Data/Images/Latent Patches"
patches_path = glob(f"{patches_dir}/*.tif")
latent_patches = []

print("[DEBUG] Reading patches.")
for path in patches_path:
  latent_patches.append(rasterio.open(path))

print("[DEBUG] Mergin mosaic.")
latent_mosaic, out_transform = merge(latent_patches)
out_meta = latent_patches[0].meta.copy()
out_meta.update({
  "height": latent_mosaic.shape[1],
  "width": latent_mosaic.shape[2],
  "transform": out_transform
})

output_dir = "./Data/Images/Latent Mosaic"
output_file = f"{output_dir}/Latent_Mosaic.tif"

print("[DEBUG] Writing file.")
os.makedirs(output_dir, exist_ok=True)
with rasterio.open(output_file, "w", **out_meta) as dst:
  dst.write(latent_mosaic)

print("[INFO] Done!")
```
