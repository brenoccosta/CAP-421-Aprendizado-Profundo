```python
# Code to transfer zipped data from Aspire5 to Quasar preserving data structure
import os
import re
from datetime import datetime

src_data = "./Data/Images Aspire5"
dst_data = "./Data/Images"
pattern = r"\sAspire5"

walkthrough = os.walk(src_data)
for root, folders, files in walkthrough:
  if files: 
    for file in files:
      src = os.path.join(root, file)
      dst = os.path.join(re.sub(pattern, "", root), file)
      os.rename(src, dst)
```
