```python
import os
import rasterio
import numpy as np

imgs_dir = "./Data/Images"
walkthrough = os.walk(imgs_dir)
output_file = "TILE.tif"

for root, folders, files in walkthrough:
  if files:
    rasters = []
    print(f"[DEBUG] {root} ...")
    
    for file in files:
      print(f"{'...':>42} {file}")
      with rasterio.open(os.path.join(root, file)) as src:
        rasters.append(src.read(1))
    
    rasters = np.array(rasters)
    out_meta = src.meta
    out_meta.update({
      "count": len(files)
    })
    
    print(f"{'':>7} {os.path.join(root, output_file)}")
    with rasterio.open(os.path.join(root, output_file), "w", **out_meta) as dst:
      dst.write(rasters)
    
    print()
    for file in files:
      os.remove(os.path.join(root, file))
```
